## Quick intro to plotting

Let's plot a $\sin(x)$ for $x = [-1,1]$

```python
npoints = 1000
x = np.linspace(-1,1,npoints)
# print(x)
```

```python
plt.plot(x, np.sin(np.pi*x))
plt.xlabel("x")
plt.ylabel("f(x)")
```

```python
def random_points_uniform(Npoints, dimension):
    rng = np.random.default_rng()

    return rng.uniform(-1,1,size=(Npoints, dimension))
```

```python
def random_points_gaussian(Npoints, dimension):
    rng = np.random.default_rng()

    return rng.normal(0,1,size=(Npoints, dimension))
```

```python
random_points_uniform(10,2)
```

```python
print(rng)
```

```python
rand_u = random_points_uniform(10000,1)
rand_g = random_points_gaussian(10000,1)
```

```python
plt.hist(rand_u)
```

```python
output = plt.hist(rand_g,bins=50)
```
